```python
from random import randint

def creat_vector(t, limi, lims):
   vect = [0] * t

   for i in range(t):
    vect[i] = random.randint(limi, lims)

   return vect
```

```python
def ordenar(vect):
  for i in range(len(vect)):
    for j in range(i+1, len(vect)):
      if(vect[i] > vect[j]):
        aux = vect[j]
        vect[j] = vect[i]
        vect[i] = aux
  return vect
```
